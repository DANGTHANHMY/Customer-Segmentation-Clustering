Trong bc tip theo, ta s to mt thuc tnh mi t **"Dt_Customer"** l **("Customer_For")** cho bit s ngy khch hng bt u mua sm ti ca hng. Cch tnh ta ly ngy ng k thnh vin ca cc khch hng so vi ngy ng k thnh vin ca khch hng gn nht.

```python
data["Dt_Customer"] = pd.to_datetime(data["Dt_Customer"])
dates = []
for i in data["Dt_Customer"]:
    i = i.date()
    dates.append(i)  
```

```python
data["Dt_Customer"] 
```

```python
#Created a feature "Customer_For"
days = []
d1 = max(dates)
for i in dates:
    delta = (d1 - i).days
    days.append(delta)
data["Customer_For"] = days
data["Customer_For"] = pd.to_numeric(data["Customer_For"], errors="coerce")
```

By gi chng ta s khm ph cc gi tr duy nht trong cc tnh nng phn loi  c  tng r rng v d liu.

```python
print("Total categories in the feature Marital_Status:\n", data["Marital_Status"].value_counts(), "\n")
print("Total categories in the feature Education:\n", data["Education"].value_counts())
```

- Thuc tnh **Marital_Status**: C 8 loi gi gi tr, khi xem xt k gng c th gom mt s gi tr li vi nhau nh: Alone, Single, widow, YOLO, Divorced, Absurb c i din bi gi tr **Alone**; Married, Together bi **Partner**. 
- Thuc tnh **Education**: c 5 loi gi tr, c th gom li thnh gi tr mi gm **Basic** i din cho Basic, **Undergraduate** i din cho 2n Cycle , **Graduate** i din cho Graduation v **Postgraduate** i din cho Master, PhD.


**Trong phn tip theo, ta s thc hin cc bc sau  thit k mt s tnh nng mi:**
- To thuc tnh **Age** ca khch hng theo **Year_Birth** cho bit nm sinh ca ngi tng ng tnh n thi im gn nht ca nm b d liu c thu thp (2015).
- To mt thuc tnh khc **Spent** cho bit tng s tin m khch hng  chi tiu trong cc danh mc khc nhau trong khong thi gian hai nm.
- To mt thuc tnh **Children**  cho bit tng s tr em trong mt h gia nh l tr em **(Kidhome)** v thanh thiu nin **(Teenhome)**. 
-  hiu r hn v h gia nh, To thuc tnh cho bit **Family_Size**, thuc tnh bng cng thc xem thuc tnh **Marital_Status** vi Alone = 1 (ngha l sng 1 mnh), Partner = 2 (ngha l c hai ngi sng chung mt nh) cng cho s ngi con ca h trong thuc tnh Children.
- To thuc tnh **Is_Parent**  hin th trng thi lm cha m, nu mt ngi c con (thuc tnh Children c gi tr t 1 tr ln), ngha l ngi ny lm cha hoc m nn gi tr **Is_Parent** s bng 1, ngc li bng 0. 
- To tnh nng **Num_Accepted** t tng s chin dch tip th trc   c khch hng chp nhn.


```python
#Feature Engineering
#Age of customer today 
data["Age"] = 2015-data["Year_Birth"]

#Total spendings on various items
data["Spent"] = data["MntWines"] + data["MntFruits"] + data["MntMeatProducts"] + data["MntFishProducts"] + data["MntSweetProducts"] + data["MntGoldProds"]

#Deriving living situation by marital status"Alone"
data["Marital_Status"]=data["Marital_Status"].replace({"Married":"Partner", "Together":"Partner", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone",})

#Feature indicating total children living in the household
data["Children"]=data["Kidhome"] + data["Teenhome"]

#Feature for total members in the householde
data["Family_Size"] =  data["Marital_Status"].replace({"Alone": 1, "Partner":2}) + data["Children"]

#Feature pertaining parenthood
data["Is_Parent"] = np.where(data.Children> 0, 1, 0)

#Segmenting education levels in three groups
data["Education"]=data["Education"].replace({"Basic":"Basic","2n Cycle":"Undergraduate", "Graduation":"Graduate", "Master":"Postgraduate", "PhD":"Postgraduate"})

#For clarity
#data=data.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold"})

# Clear description of feature
data.rename(columns={'MntWines':'Wines', 'MntFruits':'Fruits', 'MntMeatProducts':'Meats', 'MntFishProducts':'Fish', 'MntSweetProducts':'Sweets', 'MntGoldProds':'Golds'}, inplace=True)
data.rename(columns={'NumWebPurchases':'Web', 'NumCatalogPurchases':'Catalog', 'NumStorePurchases':'Store'}, inplace=True)

#Create 'Num_Accepted' feature from the sum of previous marketting campaigns that were accepted by the customer
data['Num_Accepted'] = data.AcceptedCmp1.values + data.AcceptedCmp2.values + \
                                data.AcceptedCmp3.values + data.AcceptedCmp4.values + data.AcceptedCmp5.values

#Dropping some of the redundant features
to_drop = ["Z_CostContact", "Z_Revenue", "Year_Birth", "ID","Kidhome","Teenhome","Children"]
data = data.drop(to_drop, axis=1)
print("By gi chng ta c mt s thuc tnh mi, hy xem s liu thng k ca d liu.")
data.head()
```

*	Dng LabelEncoder a cc bin  dng phn loi (categorical) v dng s (numerical).
*	S dng StandardScaler  scaling cc gi tr ca tt c cc feature.
* To dataframe con cho d liu  c gim chiu. 

```python
#Ly danh sch cc i tng categorical
s = (data.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables in the dataset:", object_cols)
```

```python
#Label Encoding
LE=LabelEncoder()
for i in object_cols:
    data[i]=data[[i]].apply(LE.fit_transform)
    
print("All features are now numerical")
```

```python
#To mt bng sao d liu
ds = data.copy()

# To mt tp hp con ca khung d liu bng cch loi b cc thuc tnh trn cc AcceptedCmp v Promotion
cols_del = ["AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3", "AcceptedCmp4", "AcceptedCmp5", "Complain", "Num_Accepted", "Response", "Dt_Customer"]
ds = ds.drop(cols_del, axis=1)

# Scaling
scaler = StandardScaler()
scaler.fit(ds)
scaled_ds = pd.DataFrame(scaler.transform(ds),columns= ds.columns )
print("All features are now scaled")
```

Biu  tng quan gia t l phng sai tch ly vi s chiu d liu:

```python
pca = PCA().fit(ds)

print(pca.explained_variance_ratio_.cumsum())

plt.plot(pca.explained_variance_ratio_.cumsum())
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```

**Gim chiu d liu vi thut ton PCA:**

T biu  scree, ta c th thy rng, ch cn t 1 thnh phn chnh tr ln c th gii thch hn 99,94% phng sai trong d liu gc, ngha l d liu c mi tng quan cao. Do  chng ta c th gim s lng bin xung t t nht 1 m vn gi li hn 99,94% thng tin c trong d liu gc.

Chng ta s thc hin gim chiu d liu cn 3. Ti v d liu 3 chiu s gip vic v v hnh dung chnh xc hn.

```python
#Initiating PCA to reduce dimentions aka features to 3
pca = PCA(n_components=3)
pca.fit(scaled_ds)
PCA_ds = pd.DataFrame(pca.transform(scaled_ds), columns=(["col1","col2", "col3"]))
PCA_ds.describe().T
```

```python
#Hnh chiu 3D ca d liu trong kch thc  gim
x =PCA_ds["col1"]
y =PCA_ds["col2"]
z =PCA_ds["col3"]
#To plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, c="maroon", marker="o" )
ax.set_title("A 3D Projection Of Data In The Reduced Dimension")
plt.show()
```

# <p style="background-color:#682F2F;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">CLUSTERING</p>

Sau khi gim chiu d liu n 3 chiu, tin hnh phn cm (cluster) bng thut ton Kmeans.

**Cc bc phn cm:**
* S dng  th Elbow  quyt nh s cm k.
* Tin hnh phn cm bng thut ton Kmeans.
* Kim tra cc cm c hnh thnh qua biu  phn tn.

```python
# Quick examination of elbow method to find numbers of clusters to make.
print('Elbow Method to determine the number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(PCA_ds)
Elbow_M.show()
```

 kim tra cc cm c hnh thnh, chng ta hy xem phn phi 3-D ca cc cm.

Chng ta s s dng k = 4, do  s c bn cm trong tp d liu ca chng ta.
Chng ta s s dng Thut ton KMeans  phn cm.

```python
km = KMeans(n_clusters = 4,random_state = 42)

y_km = km.fit_predict(PCA_ds)
PCA_ds["Clusters"] = y_km
data["Clusters"]= y_km
```

```python
#Plotting the clusters
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(x, y, z, s=40, c=PCA_ds["Clusters"], marker='o', cmap = 'RdBu_r' )
ax.set_title("The Plot Of The Clusters")
plt.show()
```

```python
y_reduced = km.fit(PCA_ds)
clusters = y_reduced.labels_

# add cluster labels to the original dataset
data['Cluster'] = clusters
data.head()
```
